```python
from imageio import imread 
from numpy.linalg import svd
from numpy import outer, shape, add, newaxis, zeros, transpose
from numpy.linalg import matrix_rank
from copy import copy, deepcopy
import matplotlib.pyplot as plt
```

```python
img = imread("/content/sample_data/Lena-Original-Image-512x512-pixels.png")
plt.imshow(img, interpolation='nearest')
plt.show()
```

```python
u, s, vh = svd(img)
r = matrix_rank(img)
m = len(u)
n = len(vh)
k = 10
u = transpose(u)
A_compressed = zeros((n, m))
for i in range(k):
    u_i = u[i]
    u_i *= s[i]
    v_i = vh[i]
    o = outer(u_i, v_i)
    A_compressed += o


plt.imshow(A_compressed, interpolation='nearest')
plt.show()
```
